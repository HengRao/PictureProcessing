```python
import sys
import os
import cv2
import numpy as np
import random
import shutil
import math

from matplotlib import pyplot as plt
from PIL import Image 
import matplotlib.patches as mpatches

path_to_video = 'Sample_Data/monkey_TaskB.mov'
%matplotlib inline
```

```python
frame_save_path = './frames_taskB/'

def create_dir_if_not_exists(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)
```

```python

cap = cv2.VideoCapture(path_to_video)
create_dir_if_not_exists(frame_save_path) # Or you can create it manully
if not cap.isOpened():
    print('{} not opened'.format(path_to_video))
    sys.exit(1)
time_length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
frame_height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
frame_width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
frame_counter = 0                                             # FRAME_COUNTER
while(1):
    return_flag, frame = cap.read()  
    if not return_flag:
        print('Video Reach End')
        break
    # Main Content - Start
    cv2.imshow('', frame)
    cv2.imwrite(frame_save_path + 'frame%d.tif' % frame_counter, frame)
    frame_counter += 1
    # Main Content - End    
    if cv2.waitKey(30) & 0xff == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

```python
def drawbox(frame, centerx, centery, radius, color):
    for y in range(centerx - radius, centerx + radius):
        for x in range(centery - radius, centery + radius):
            cx = 0 if x < 0 else frame.shape[0]-1 if x > frame.shape[0] - 1 else x
            cy = 0 if y < 0 else frame.shape[1]-1 if y > frame.shape[1] - 1 else y
            for i in range(3):
                frame[cx][cy][i] = color[i]
    return frame



BLUE = 120
# count = 193
bgctr = 193
count = 0

RED = 255
cap = cv2.VideoCapture('Sample_Data/monkey_TaskB.mov')
if not cap.isOpened():
    print('monkey.avi not opened')
    sys.exit(1)
length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
frame_height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
frame_width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
while(1):
    ret, monkeyframe = cap.read()
    if not ret:
        break
    
    bg = cv2.imread('background/frame%d.tif' % (count%bgctr))
    bg1 = cv2.imread('background/frame%d.tif' % (count%bgctr))
    if bg is None:
        print('ooops! no bg found BG/frame%d.tif' % (count%bgctr))
        break
    # overwrite the background
    
    monkeyframe = cv2.resize(monkeyframe,(bg.shape[1],bg.shape[0]),interpolation = cv2.INTER_CUBIC)
 
   
    
    for x in range(monkeyframe.shape[0]):
        for y in range(monkeyframe.shape[1]):
            if monkeyframe[x,y,0] < 110 and  monkeyframe[x,y,1]<190 and monkeyframe[x,y,2]>190 and monkeyframe[x,y,2]<=255 :                
                bg[x,y] = 255   

            else:
                bg[x,y] = 0
                
                   
    kernel = np.ones((3,3), np.uint8)
    
    bg = cv2.erode(bg, kernel, iterations=1)
    bg = cv2.dilate(bg, kernel, iterations=4) 
    #bg = cv2.erode(bg, kernel, iterations=1)
    
    

    print()
     
    points = []
    print(bg.shape)    
    
    
    for x in range(bg.shape[0]):     
        for y in range(bg.shape[1]):
            if bg[x,y,0] == 255:
                points.append((x,y))
            
    
    
    x1 = points[0][1]
    y1 = points[0][0]
    
    x2 = points[int(len(points)/2)][1]
    y2 = points[int(len(points)/2)][0]
    
    bg1 = cv2.circle(bg1, (x2, y2), 20, (255,255,255), -1)
    
    bg1 = cv2.circle(bg1, (x1, y1), 10, (255,255,255), -1)     
    bg1 = cv2.line(bg1, (x1, y1), (x2, y2), (255, 255, 255), 2)
  

    x1 = points[int(len(points)/4)][1]
    y1 = points[int(len(points)/4)][0]
    
    bg1 = cv2.circle(bg1, (x1, y1), 10, (255,255,255), -1)
    bg1 = cv2.line(bg1, (x1, y1), (x2, y2), (255, 255, 255), 2)
  

    x1 = points[int(len(points)*3/4)][1]
    y1 = points[int(len(points)*3/4)][0]
    bg1 = cv2.circle(bg1, (x1, y1), 10, (255,255,255), -1)
    bg1 = cv2.line(bg1, (x1, y1), (x2, y2), (255, 255, 255), 2)
  
    
    x1 = points[len(points)-1][1]
    y1 = points[len(points)-1][0]
    bg1 = cv2.circle(bg1, (x1, y1), 10, (255,255,255), -1)
    bg1 = cv2.line(bg1, (x1, y1), (x2, y2), (255, 255, 255), 2)
    
    #image = cv2.circle(bg, center_coordinates, radius, color, thickness) 
 
          
                    
    cv2.imwrite('composite_taskB/composite%d.tif' % count, bg1)
    cv2.putText(img=bg1, text='Compositing: %d%%' % int(100*count/length), org=(int(0), int(bg1.shape[1] / 2)),
                fontFace=cv2.FONT_HERSHEY_DUPLEX, fontScale=0.7,
                color=(0, 255, 0))
    cv2.imshow('MonkeyB', bg1)

    count += 1
    if cv2.waitKey(30) & 0xff == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()

```
